```python
%matplotlib notebook

import matplotlib.pyplot
from PIL import Image
import numpy

image = Image.open("20200522_213922.jpg")
grey_image = image.convert("L")

numpy_image = numpy.asarray(grey_image)
# 800, 800+256*5
# 1500, 1500+256*5
numpy_barcode = numpy.array(numpy_image[800:2080, 1500:2780]).astype(float)
numpy_barcode -= sum(sum(numpy_barcode)) / numpy_barcode.size

figure = matplotlib.pyplot.figure()
axes = figure.add_subplot(1,1,1)

axes.imshow(numpy_barcode)

figure.show()

barcode_f = numpy.fft.fft2(numpy_barcode)

figure = matplotlib.pyplot.figure()
axes = figure.add_subplot(1,1,1)

axes.imshow(numpy.fft.fftshift(abs(barcode_f[1:-1,1:-1])),
            extent=(-640,639,-640,639), origin='lower')
#axes.set_xlim(0,50)
#axes.set_ylim(0,50)

figure.show()
```
